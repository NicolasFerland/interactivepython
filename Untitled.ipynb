```python
# Exercice with fraction class

# http://interactivepython.org/runestone/static/pythonds/Introduction/ObjectOrientedProgramminginPythonDefiningClasses.html#a-fraction-class
# http://interactivepython.org/runestone/static/pythonds/Introduction/ProgrammingExercises.html


def gcd(m,n):
    while m%n != 0:
        oldm = m
        oldn = n

        m = oldn
        n = oldm%oldn
    return n

class Fraction:
     def __init__(self,top,bottom):
         if not (isinstance(top, int) and isinstance(bottom, int)) :
            raise RuntimeError("Error: Fraction class only accepts integer")
         if bottom < 0:
            top, bottom = -top, -bottom # Put the minus sign on the numerator
         common = gcd(top,bottom)
         self.num = top // common
         self.den = bottom // common

     def __str__(self):
        if self.den != 1:
            return str(self.num)+"/"+str(self.den)
        else:
            return str(self.num)

     def show(self):
         print(self.num,"/",self.den)

     def __add__(self,otherfraction):
         if not isinstance(otherfraction, Fraction) :
                otherfraction = Fraction(otherfraction,1)
         newnum = self.num*otherfraction.den + \
                      self.den*otherfraction.num
         newden = self.den * otherfraction.den
         return Fraction(newnum,newden)
        
     # def __eq__(self, other):
     #    firstnum = self.num * other.den
     #    secondnum = other.num * self.den
     #    return firstnum == secondnum
        
     def getNum(self):
        return self.num
        
     def getDen(self):
        return self.den
        
     def __sub__(self,otherfraction):
        if not isinstance(otherfraction, Fraction) :
            otherfraction = Fraction(otherfraction,1)
        newnum = self.num*otherfraction.den - \
                      self.den*otherfraction.num
        newden = self.den * otherfraction.den
        return Fraction(newnum,newden)
        
     def __mul__(self,otherfraction):
        if not isinstance(otherfraction, Fraction) :
            otherfraction = Fraction(otherfraction,1)
        newnum = self.num*otherfraction.num
        newden = self.den * otherfraction.den
        return Fraction(newnum,newden)
        
     def __truediv__(self,otherfraction):
        if not isinstance(otherfraction, Fraction) :
            otherfraction = Fraction(otherfraction,1)
        newnum = self.num*otherfraction.den
        newden = self.den * otherfraction.num
        return Fraction(newnum,newden)
    
     def __eq__(self, other):
        if not isinstance(other, Fraction) :
            other = Fraction(other,1)
        return self.num == other.num and self.den == other.den
    
     def __ne__(self, other):
        if not isinstance(other, Fraction) :
            other = Fraction(other,1)
        return self.num != other.num or self.den != other.den
    
     def __gt__(self, other):
        if not isinstance(other, Fraction) :
            other = Fraction(other,1)
        firstnum = self.num * other.den
        secondnum = other.num * self.den
        return firstnum > secondnum
    
     def __ge__(self, other):
        if not isinstance(other, Fraction) :
            other = Fraction(other,1)
        firstnum = self.num * other.den
        secondnum = other.num * self.den
        return firstnum >= secondnum
    
     def __lt__(self, other):
        if not isinstance(other, Fraction) :
            other = Fraction(other,1)
        firstnum = self.num * other.den
        secondnum = other.num * self.den
        return firstnum < secondnum
    
     def __le__(self, other):
        if not isinstance(other, Fraction) :
            other = Fraction(other,1)
        firstnum = self.num * other.den
        secondnum = other.num * self.den
        return firstnum <= secondnum
    
    # define r operator to reverse integer and fraction.
     def __radd__(self,otherfraction):
        if not isinstance(otherfraction, Fraction) :
                otherfraction = Fraction(otherfraction,1)
        newnum = self.num*otherfraction.den + \
                      self.den*otherfraction.num
        newden = self.den * otherfraction.den
        return Fraction(newnum,newden)
     
     def __rsub__(self,otherfraction):
        if not isinstance(otherfraction, Fraction) :
            otherfraction = Fraction(otherfraction,1)
        newnum = self.num*otherfraction.den - \
                      self.den*otherfraction.num
        newden = self.den * otherfraction.den
        return Fraction(-newnum,newden)
        
     def __rmul__(self,otherfraction):
        if not isinstance(otherfraction, Fraction) :
            otherfraction = Fraction(otherfraction,1)
        newnum = self.num*otherfraction.num
        newden = self.den * otherfraction.den
        return Fraction(newnum,newden)
        
     def __rtruediv__(self,otherfraction):
        if not isinstance(otherfraction, Fraction) :
            otherfraction = Fraction(otherfraction,1)
        newnum = self.num*otherfraction.den
        newden = self.den * otherfraction.num
        return Fraction(newden,newnum)
    
     def __req__(self, other):
        if not isinstance(other, Fraction) :
            other = Fraction(other,1)
        return self.num == other.num and self.den == other.den
    
     def __rne__(self, other):
        if not isinstance(other, Fraction) :
            other = Fraction(other,1)
        return self.num != other.num or self.den != other.den
    
     def __rgt__(self, other):
        if not isinstance(other, Fraction) :
            other = Fraction(other,1)
        firstnum = self.num * other.den
        secondnum = other.num * self.den
        return firstnum > secondnum
    
     def __rge__(self, other):
        if not isinstance(other, Fraction) :
            other = Fraction(other,1)
        firstnum = self.num * other.den
        secondnum = other.num * self.den
        return firstnum >= secondnum
    
     def __rlt__(self, other):
        if not isinstance(other, Fraction) :
            other = Fraction(other,1)
        firstnum = self.num * other.den
        secondnum = other.num * self.den
        return firstnum < secondnum
    
     def __rle__(self, other):
        if not isinstance(other, Fraction) :
            other = Fraction(other,1)
        firstnum = self.num * other.den
        secondnum = other.num * self.den
        return firstnum <= secondnum
    
     def __repr__(self):
        return "Fraction(%r,%r)" % (self.num,self.den)


x = Fraction(1,2)
y = Fraction(2,3)
print(x+y)
print(x+x+y)
print(x+x+y == Fraction(5,3))
print(x+x+y == Fraction(10,6))
print(x+x+y == Fraction(5,2))
print(y.getNum())
print(x + 2)
print(x * y)
print(x * 2)
print(x / 2)
print(2 + x)
print(2 * x)
print(2 - x)
print(2 / x)
print(2 > x)
print(2 == x)
print(2 < x)
print(0 > x)
y += Fraction(4,5)
print(y)
print(repr(y))
print(eval(repr(y))==y)
```

```python
# Recursive function

# Example to convert number to string with some base

def toStr(n,base):
   convertString = "0123456789ABCDEF" # This example can work for 2, 10, 16 or any other 1<base<16
   if n < base:
      return convertString[n]
   else:
      return toStr(n//base,base) + convertString[n%base]

print(toStr(1453,16))
print(toStr(1453,10))
print(toStr(1453,2))
```

```python
# Write a function that takes a string as a parameter and returns True if the string is a palindrome
import string
table = dict.fromkeys(map(ord, string.punctuation))
table[32] = None # Space is 32. Somehow table doesn't work for apostrophe even though 39 is included
#table[39] = None   It's because ’ is not part of ASCII and is not '.
def isPalindrome(word):
    word = word.translate(table).lower().replace("’","")
    return isPalindrome0(word)

def isPalindrome0(word):
    end = len(word) - 1
    if end < 2: # end = 0 or 1, so the length is 1 or 2
        return word[0] == word[end]
    return word[0] == word[end] and isPalindrome0(word[1:end])

print("Should be true")
print(isPalindrome("aibohphobia"))
print(isPalindrome("Live not on evil"))
print(isPalindrome("Reviled did I live, said I, as evil I did deliver"))
print(isPalindrome("Go hang a salami; I’m a lasagna hog."))

print("Should be false")
print(isPalindrome("aibocphobia"))
print(isPalindrome("ajvg"))
```

```python
# They say to use dynamic programming to do problem 15.
# Let's look at their example of dynamic programing

def dpMakeChange(coinValueList,change,minCoins,coinsUsed): # minCoins, coinsUsed are zero arrays
    # coinValueList is the list of allowed coins: [1,5,10,25]
    # change is the total value we want to have
   for cents in range(change+1): # Loop from 0 to change
      coinCount = cents # coinCount if using only 1s
      newCoin = 1
      for j in [c for c in coinValueList if c <= cents]: # Loop over all coins < change
            # minCoins[cents-j] + 1 is coinCount if using one coin of value j + a previous solution
            # we compare this solution to using only 1s, if it's better we save it.
            # after we compare new solutions to the one we saved, and replace it if it beats it.
            if minCoins[cents-j] + 1 < coinCount:
               coinCount = minCoins[cents-j]+1
               newCoin = j
      minCoins[cents] = coinCount # Save the min number of Coins for this value of cents.
      coinsUsed[cents] = newCoin # Save the new coin for this value of cents.
        # The other coins used for this value of cents can be found by finding all coinsUsed value
        # while decreasing cents by the value of used coin.
   return minCoins[change]

def printCoins(coinsUsed,change):
   coin = change
   while coin > 0:
      thisCoin = coinsUsed[coin]
      print(thisCoin)
      coin = coin - thisCoin

def main():
    amnt = 63
    clist = [1,5,10,25]
    coinsUsed = [0]*(amnt+1)
    coinCount = [0]*(amnt+1)

    print("Making change for",amnt,"requires")
    print(dpMakeChange(clist,amnt,coinCount,coinsUsed),"coins")
    print("They are:")
    printCoins(coinsUsed,amnt)

main()
```

```python
# Problem 15  string edit distance problem

# Dynamic programming is to split the problem in many small ones and reuse the previous solutions.
# The small problem is to do one letter at a time. For each letter we have three options:
# replace, remove, insert. We try the 3 options and do the small problem again.

def editdistance(word, target, m, n):
    ed = numpy.zeros(shape=(m+1,n+1)).astype(int)
    for i in range(0, m+1): # 0: before word, 1 to m : each letter
        for j in range(0, n+1):
            if i == 0: # word is finished, insert what's remain of the target
                ed[i,j] = 20*j
            elif j == 0: # target is finished, remove what's remain of the word
                ed[i,j] = 20*i
            elif word[i-1] == target[j-1]: # same letter, so go to the next letter
                ed[i,j] = ed[i-1,j-1]
            else: # find min of 3 options
                ed[i,j]=min(
                    5 + ed[i-1,j-1],  # replace, then go to previous letter
                    20 + ed[i-1,j], # remove letter, then compare with previous letter
                    20 + ed[i,j-1] # insert target, then compare same letter with previous target  
                )
    return ed[m,n]
        
word = "algorithm"
target = "alligator"
print("Transforming", word, "into",target,"requires a cost of",
      editdistance(word, target, len(word), len(target)))
```
