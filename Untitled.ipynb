```python
# Exercice with fraction class

# http://interactivepython.org/runestone/static/pythonds/Introduction/ObjectOrientedProgramminginPythonDefiningClasses.html#a-fraction-class
# http://interactivepython.org/runestone/static/pythonds/Introduction/ProgrammingExercises.html


def gcd(m,n):
    while m%n != 0:
        oldm = m
        oldn = n

        m = oldn
        n = oldm%oldn
    return n

class Fraction:
     def __init__(self,top,bottom):
         if not (isinstance(top, int) and isinstance(bottom, int)) :
            raise RuntimeError("Error: Fraction class only accepts integer")
         if bottom < 0:
            top, bottom = -top, -bottom # Put the minus sign on the numerator
         common = gcd(top,bottom)
         self.num = top // common
         self.den = bottom // common

     def __str__(self):
        if self.den != 1:
            return str(self.num)+"/"+str(self.den)
        else:
            return str(self.num)

     def show(self):
         print(self.num,"/",self.den)

     def __add__(self,otherfraction):
         if not isinstance(otherfraction, Fraction) :
                otherfraction = Fraction(otherfraction,1)
         newnum = self.num*otherfraction.den + \
                      self.den*otherfraction.num
         newden = self.den * otherfraction.den
         return Fraction(newnum,newden)
        
     # def __eq__(self, other):
     #    firstnum = self.num * other.den
     #    secondnum = other.num * self.den
     #    return firstnum == secondnum
        
     def getNum(self):
        return self.num
        
     def getDen(self):
        return self.den
        
     def __sub__(self,otherfraction):
        if not isinstance(otherfraction, Fraction) :
            otherfraction = Fraction(otherfraction,1)
        newnum = self.num*otherfraction.den - \
                      self.den*otherfraction.num
        newden = self.den * otherfraction.den
        return Fraction(newnum,newden)
        
     def __mul__(self,otherfraction):
        if not isinstance(otherfraction, Fraction) :
            otherfraction = Fraction(otherfraction,1)
        newnum = self.num*otherfraction.num
        newden = self.den * otherfraction.den
        return Fraction(newnum,newden)
        
     def __truediv__(self,otherfraction):
        if not isinstance(otherfraction, Fraction) :
            otherfraction = Fraction(otherfraction,1)
        newnum = self.num*otherfraction.den
        newden = self.den * otherfraction.num
        return Fraction(newnum,newden)
    
     def __eq__(self, other):
        if not isinstance(other, Fraction) :
            other = Fraction(other,1)
        return self.num == other.num and self.den == other.den
    
     def __ne__(self, other):
        if not isinstance(other, Fraction) :
            other = Fraction(other,1)
        return self.num != other.num or self.den != other.den
    
     def __gt__(self, other):
        if not isinstance(other, Fraction) :
            other = Fraction(other,1)
        firstnum = self.num * other.den
        secondnum = other.num * self.den
        return firstnum > secondnum
    
     def __ge__(self, other):
        if not isinstance(other, Fraction) :
            other = Fraction(other,1)
        firstnum = self.num * other.den
        secondnum = other.num * self.den
        return firstnum >= secondnum
    
     def __lt__(self, other):
        if not isinstance(other, Fraction) :
            other = Fraction(other,1)
        firstnum = self.num * other.den
        secondnum = other.num * self.den
        return firstnum < secondnum
    
     def __le__(self, other):
        if not isinstance(other, Fraction) :
            other = Fraction(other,1)
        firstnum = self.num * other.den
        secondnum = other.num * self.den
        return firstnum <= secondnum
    
    # define r operator to reverse integer and fraction.
     def __radd__(self,otherfraction):
        if not isinstance(otherfraction, Fraction) :
                otherfraction = Fraction(otherfraction,1)
        newnum = self.num*otherfraction.den + \
                      self.den*otherfraction.num
        newden = self.den * otherfraction.den
        return Fraction(newnum,newden)
     
     def __rsub__(self,otherfraction):
        if not isinstance(otherfraction, Fraction) :
            otherfraction = Fraction(otherfraction,1)
        newnum = self.num*otherfraction.den - \
                      self.den*otherfraction.num
        newden = self.den * otherfraction.den
        return Fraction(-newnum,newden)
        
     def __rmul__(self,otherfraction):
        if not isinstance(otherfraction, Fraction) :
            otherfraction = Fraction(otherfraction,1)
        newnum = self.num*otherfraction.num
        newden = self.den * otherfraction.den
        return Fraction(newnum,newden)
        
     def __rtruediv__(self,otherfraction):
        if not isinstance(otherfraction, Fraction) :
            otherfraction = Fraction(otherfraction,1)
        newnum = self.num*otherfraction.den
        newden = self.den * otherfraction.num
        return Fraction(newden,newnum)
    
     def __req__(self, other):
        if not isinstance(other, Fraction) :
            other = Fraction(other,1)
        return self.num == other.num and self.den == other.den
    
     def __rne__(self, other):
        if not isinstance(other, Fraction) :
            other = Fraction(other,1)
        return self.num != other.num or self.den != other.den
    
     def __rgt__(self, other):
        if not isinstance(other, Fraction) :
            other = Fraction(other,1)
        firstnum = self.num * other.den
        secondnum = other.num * self.den
        return firstnum > secondnum
    
     def __rge__(self, other):
        if not isinstance(other, Fraction) :
            other = Fraction(other,1)
        firstnum = self.num * other.den
        secondnum = other.num * self.den
        return firstnum >= secondnum
    
     def __rlt__(self, other):
        if not isinstance(other, Fraction) :
            other = Fraction(other,1)
        firstnum = self.num * other.den
        secondnum = other.num * self.den
        return firstnum < secondnum
    
     def __rle__(self, other):
        if not isinstance(other, Fraction) :
            other = Fraction(other,1)
        firstnum = self.num * other.den
        secondnum = other.num * self.den
        return firstnum <= secondnum
    
     def __repr__(self):
        return "Fraction(%r,%r)" % (self.num,self.den)


x = Fraction(1,2)
y = Fraction(2,3)
print(x+y)
print(x+x+y)
print(x+x+y == Fraction(5,3))
print(x+x+y == Fraction(10,6))
print(x+x+y == Fraction(5,2))
print(y.getNum())
print(x + 2)
print(x * y)
print(x * 2)
print(x / 2)
print(2 + x)
print(2 * x)
print(2 - x)
print(2 / x)
print(2 > x)
print(2 == x)
print(2 < x)
print(0 > x)
y += Fraction(4,5)
print(y)
print(repr(y))
print(eval(repr(y))==y)
```

```python
# Recursive function

# Example to convert number to string with some base

def toStr(n,base):
   convertString = "0123456789ABCDEF" # This example can work for 2, 10, 16 or any other 1<base<16
   if n < base:
      return convertString[n]
   else:
      return toStr(n//base,base) + convertString[n%base]

print(toStr(1453,16))
print(toStr(1453,10))
print(toStr(1453,2))
```

```python
# Write a function that takes a string as a parameter and returns True if the string is a palindrome
import string
table = dict.fromkeys(map(ord, string.punctuation))
table[32] = None # Space is 32. Somehow table doesn't work for apostrophe even though 39 is included
#table[39] = None   It's because ’ is not part of ASCII and is not '.
def isPalindrome(word):
    word = word.translate(table).lower().replace("’","")
    return isPalindrome0(word)

def isPalindrome0(word):
    end = len(word) - 1
    if end < 2: # end = 0 or 1, so the length is 1 or 2
        return word[0] == word[end]
    return word[0] == word[end] and isPalindrome0(word[1:end])

print("Should be true")
print(isPalindrome("aibohphobia"))
print(isPalindrome("Live not on evil"))
print(isPalindrome("Reviled did I live, said I, as evil I did deliver"))
print(isPalindrome("Go hang a salami; I’m a lasagna hog."))

print("Should be false")
print(isPalindrome("aibocphobia"))
print(isPalindrome("ajvg"))
```

```python
# Do some of those problems:
# http://interactivepython.org/runestone/static/pythonds/Recursion/pythondsProgrammingExercises.html

# Problem 9-10  2 jugs problem
import numpy
Step=[]
largeSize = 4
smallSize = 3
final = 2
Visited = numpy.zeros(shape=(smallSize+1,largeSize+1)) # 0 is True
def jugs(initialSmall, initialLarge):
    
    if Visited[initialSmall, initialLarge]:
        return False # It is never useful to go back where we went
    
    print(initialSmall, initialLarge)

    if initialLarge == final:
        return True # That means the problem is solved
    
    Visited[initialSmall, initialLarge] = True
    
    if jugs(initialSmall, largeSize): # Fill large
        Step.append(1)
        return True
    if jugs(smallSize, initialLarge): # Fill small
        Step.append(2)
        return True
    if jugs(min(smallSize,initialLarge+initialSmall), 
            max(initialLarge-(smallSize-initialSmall),0)):# Empty large into small
        Step.append(3)
        return True
    if jugs(max(initialSmall-(largeSize-initialLarge),0), 
            min(largeSize,initialLarge+initialSmall)):# Empty small into large
        Step.append(4)
        return True
    if jugs(initialSmall, 0):# Empty large
        Step.append(5)
        return True
    if jugs(0, initialLarge):# Empty small
        Step.append(6)
        return True
    
    return False
    
print(jugs(0,0))    
print(Step[::-1])

# Filling then emptying large was useless, could have had: [2, 4, 2, 4, 5, 4]
```

```python
# Try to do it so that it's always the best

from functools import reduce

def FindShortestList(x,y):
    if len(x) < len(y):
        return x
    else:
        return y

def jugs(initialSmall, initialLarge):
    global Steps
    global Step
    
    if Visited[initialSmall, initialLarge]:
        return False # It is never useful to go back where we went

    if initialLarge == final:
        if Step != []:
            Steps.append(list(Step))
        Step = []
        return True # That means the problem is solved
    
    Visited[initialSmall, initialLarge] = True
    Solved=0
    
    if jugs(initialSmall, largeSize): # Fill large
        Step.append(1)
        Solved += 1
        Visited[initialSmall, largeSize] = False # Wants to be able to go back there if there is another solution
    if jugs(smallSize, initialLarge): # Fill small
        Step.append(2)
        Solved += 1
        Visited[smallSize, initialLarge] = False 
    if jugs(min(smallSize,initialLarge+initialSmall), 
            max(initialLarge-(smallSize-initialSmall),0)):# Empty large into small
        Step.append(3)
        Solved += 1
        Visited[min(smallSize,initialLarge+initialSmall), 
            max(initialLarge-(smallSize-initialSmall),0)] = False 
    if jugs(max(initialSmall-(largeSize-initialLarge),0), 
            min(largeSize,initialLarge+initialSmall)):# Empty small into large
        Step.append(4)
        Solved += 1
        Visited[max(initialSmall-(largeSize-initialLarge),0), 
            min(largeSize,initialLarge+initialSmall)] = False 
    if jugs(initialSmall, 0):# Empty large
        Step.append(5)
        Solved += 1
        Visited[initialSmall, 0] = False 
    if jugs(0, initialLarge):# Empty small
        Step.append(6)
        Solved += 1
        Visited[0, initialLarge] = False 
    
    if Solved > 1: # If 2 or more at the same level have been solved, that means we want the smallest
        # between them.
        Steps.append(list(Step))
        Step = reduce(FindShortestList,Steps[-Solved:]) 
        if Solved == len(Steps):
            Steps = []
        else:
            Steps = Steps[:-Solved] # We want to keep any step before the last #Solved ones since they are
        # at the higher level.
    
    return Solved > 0

Steps=[]
Step=[]
largeSize = 4
smallSize = 3
final = 2
Visited = numpy.zeros(shape=(smallSize+1,largeSize+1))

jugs(0,0)
print(Step[::-1])
```
